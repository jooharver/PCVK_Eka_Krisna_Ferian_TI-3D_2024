```python
#Tugas 2 Kelompok

from google.colab import drive
# Accessing Google Drive
drive.mount('/content/drive')
import cv2 as cv
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

# Step 1
img = cv.imread('/content/drive/MyDrive/PCVK/images/KTP7.jpeg')
# Convert BGR to RGB
img_ktp = cv.cvtColor(img, cv.COLOR_BGR2RGB)
cv.rectangle(img_ktp, (190 , 70), (535, 98), (0, 255, 0), -1)  # Rectangle untuk NIK
cv.rectangle(img_ktp, (190 , 120), (500, 140), (255,0 , 0), -1)  # Rectangle untuk Nama
cv.circle(img_ktp, (641, 198), 15, (255, 0, 0), 3) #circle mata kiri
cv.circle(img_ktp, (693, 198), 15, (0, 255, 0), 3) #circle mata kanan
cv.circle(img_ktp, (667, 255), 15, (0, 0, 255), 3) #circle mulut
# Display the image
plt.imshow(img_ktp)
```
